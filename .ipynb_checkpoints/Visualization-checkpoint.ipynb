```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('Titanic.csv')
df = df.drop('Unnamed: 0',axis=1)
df['Sex'] = df['Sex'].replace({1:'Male',0:'Female'})
df['Embarked'] = df['Embarked'].replace({1:'Queensland',0:'Cherbourg',2:'Southampton'})

df1 = pd.read_csv('Titanic.csv')
df1 = df1.drop('Unnamed: 0',axis=1)
df1['Sex'] = df1['Sex'].replace({1:'Male',0:'Female'})
df1['Embarked'] = df1['Embarked'].replace({1:'Queensland',0:'Cherbourg',2:'Southampton'})
```

```python
plt.figure(figsize=(15,10))
plt.subplot(231)
sb.set(style='whitegrid')
sb.barplot(data=df,x='Sex',y='Survived',edgecolor=(0,0,0),linewidth=2, ci=None)

plt.subplot(232)
sb.set(style='whitegrid')
sb.barplot(data=df,x='Pclass',y='Survived',edgecolor=(0,0,0),linewidth=2, ci=None)

plt.subplot(233)
sb.set(style='whitegrid')
sb.barplot(data=df,x='SibSp',y='Survived',edgecolor=(0,0,0),linewidth=2, ci=None)

plt.subplot(234)
sb.set(style='whitegrid')
sb.barplot(data=df,x='Parch',y='Survived',edgecolor=(0,0,0),linewidth=2, ci=None)

plt.subplot(235)
sb.set(style='whitegrid')
sb.barplot(data=df,x='Embarked',y='Survived',edgecolor=(0,0,0),linewidth=2, ci=None)

# plt.savefig('all.jpg')
```

```python
plt.figure(figsize=(15,10))
plt.subplot(231)
sb.set(style='whitegrid')
sb.countplot(df['Sex'],edgecolor=(0,0,0),linewidth=2)

plt.subplot(232)
sb.set(style='whitegrid')
sb.countplot(df['Pclass'],edgecolor=(0,0,0),linewidth=2)

plt.subplot(233)
sb.set(style='whitegrid')
sb.countplot(df['SibSp'],edgecolor=(0,0,0),linewidth=2)

plt.subplot(234)
sb.set(style='whitegrid')
sb.countplot(df['Parch'],edgecolor=(0,0,0),linewidth=2)

plt.subplot(235)
sb.set(style='whitegrid')
sb.countplot(df['Embarked'],edgecolor=(0,0,0),linewidth=2)

# plt.savefig('count.jpg')
```
